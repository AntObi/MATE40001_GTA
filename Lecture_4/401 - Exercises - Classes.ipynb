```python
# Make a new class and call it Animal.
# Animal should have an attribute "name" and "age"

# Use the class Animal to define a derived class Dog. The class Dog has two additional method attributes
# "myName", which print out the value of the data attribute name
# "humanYears", that takes the age of the dog in the data attribute age, multiplies it by 7 and prints it out

class Animal():
    def __init__(self, name, age):
        self.name=name
        self.age = age
        return
    

class Dog( Animal ):
    def myName(self):
        return self.name
        
    
    def humanYears(self):
        return self.age*7
        

# If you do it correctly, Running the next cells should give you the output:
# 'My name is Doggo and I am 49 years old'
```

```python
dog = Dog( "Doggo", 7 )
print( "My name is {0} and I am {1} human-years old".format( dog.myName(), dog.humanYears() ) ) 
```

```python
# Define a new class that is a list (so it is a derived class which inherits from the "list" object) 
# with some special methods.

# call your class superList
 
# the class has the two method attributes:
# difference: this method takes as input another list (call it list1). If the list is of the same length, than it return
#             a list whose entries are the element by element difference of "self" and "list1". For example, if you
#             listA is an instance of the class superList and has been initialised with the values [1,2,3] 
#             writing listA.difference[ 1, 2, 4 ] should return [ 0, 0, -1 ]. If not, an Error is raised

# multiply: as in "difference above, but a multiplication is made. In this case, listA.multiply[ 1, 2, 4 ] 
#           should return [ 1*1 = 1, 2*2 = 4, 3*4 = 12  ]. If of different lengths, an Error is raised

class superList( list ):
    def difference(self, list1):
        if len(self)!=len(list1):
            print("Error, the lists are not the same length")
            raise ValueError
        diff=[]
        for i in range(len(self)):
            diff.append(self[i]-list1[i])
        return diff
    
    def multiply(self, list1):
        if len(self)!=len(list1):
            print("Error, the lists are not the same length")
            raise ValueError
        diff=[]
        for i in range(len(self)):
            diff.append(self[i]*list1[i])
        return diff
        

#If correct, running the next cell you should obtain the following output:
# [0, 0, 0, 0, 0]
# [1, 4, 9, 16, 25]
                
```
