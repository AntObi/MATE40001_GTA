```python
# Define a new class that is a list (so it is a derived class which inherits from the "list" object) 
# with some special methods.

# call your class superList
 
# the class has the two method attributes:
# difference: this method takes as input another list (call it list1). If the list is of the same length, than it return
#             a list whose entries are the element by element difference of "self" and "list1". For example, if you
#             listA is an instance of the class superList and has been initialised with the values [1,2,3] 
#             writing listA.difference[ 1, 2, 4 ] should return [ 0, 0, -1 ]. If not, an Error is raised

# multiply: as in "difference above, but a multiplication is made. In this case, listA.multiply[ 1, 2, 4 ] 
#           should return [ 1*1 = 1, 2*2 = 4, 3*4 = 12  ]. If of different lengths, an Error is raised

class superList( list ):
    def difference(self, list1):
        if len(self) != len(list1):
            print('Error: the two lists should be of the same length')
            raise ValueError
        differencelist = []
        for i in range(len(list1)):
            differencelist.append(self[i] - list1[i])
        return differencelist
    def multiply(self, list1):
        if len(self) != len(list1):
            print('Error: the two lists should be of the same length')
            raise ValueError
        multipliedlist = []
        for i in range(len(list1)):
            multipliedlist.append(self[i] * list1[i])
        return multipliedlist

#If correct, running the next cell you should obtain the following output:
# [0, 0, 0, 0, 0]
# [1, 4, 9, 16, 25]
                
```

```python
# Define a new type of dictionary, call this class "specialDictionary". 
# This class should have a method that you should call "exchange". 
# This method takes as input two arbitrary characters and change each of the values 
# present in the dictionary, which should all be strings, and exchange the first
# of the given input characters, if present, with the second. 
# For example, if we give as input the characters "a" and "e" and the words 
# "Lalaland" and "massage" are present in the dictionary values, they should become 
# "Lelelend" and "messege".

# Bear in mind that items in a dictionary are TUPLES ( = immutable objects... ) of key:value 
# pairs

class specialDictionary( dict ):

    def exchange(self, old, new):
        # check all values are of type str
        for val in self.values():
            if type(val)!=str:
                print('Error: all values need to be strings.')
                raise ValueError
        for key in self.keys():
            self[key] = self[key].replace(old, new)
        print(self)
        return
        
```

```python
# Define a new class called Day. 
# Day have an attribute "hours" which is a dictionary where keys are the hours of the day, and value
# is a string with a description of the activity in that hour, or the value None.
# Upon declaration, "hours" should contain only None values.
# Day should have two methods:
# clearDay: Remove from "hours" all the activities for that day. clearDay does not take any input.
# addActivity: Look into "hours" and checks. If no activity is present at that time, add a string 
#              with a description of the activity for the day, otherwise return the message
#              "You are already busy!". addActivity takes two inputs, a specific hour (an integer
# .            from 0 to 23 ) and a string with the activity description
# removeActivity: Look into "hours" and checks. If an activity is present at the specified hour, it is removed
#              if not, nothing happens. removeActivity takes a single input, a specific hour (an integer
# .            from 0 to 23 )

class Day( ):
    def __init__(self):
        self.hours = dict()
        for i in range(24):
            self.hours[i] = None
        return
    def clearDay(self):
        for key in self.hours.keys():
            self.hours[key] = None
        return
    def addActivity(self, hour, activity):
        if self.hours[hour]:
            print('You are already busy!')
            return
        else:
            self.hours[hour] = activity
        return
    def removeActivity(self, hour):
        self.hours[hour] = None
        return
```

```python
# Let us now define a new class called yearAgenda.
# This class should contain an attribute "Days" which is a list of objects of class Day.
# Days should be initially empty.

# somehow similarly to Day, yearAgenda should have the following methods:
## clearDay: Remove all activities from that day. clearDay takes as input a number (1-365), the number of the 
#            day in the year.
# addActivity: Look into a specific value of Days at a specific value of "hours" and checks. 
#              If no activity is present at that time, add a string 
#              with a description of the activity for the day and hour, otherwise return the message
#              "You are already busy!". addActivity takes three inputs, a specific day number, a specific 
#              hour (an integer from 0 to 23 ) and a string with the activity description.
# removeActivity: Look into Days at a specific value of hours. If an activity is present at the specified hour, 
#              it is removed, otherwise nothing happens. 
#              removeActivity takes two input, a day number (1-365) and an hour number ( 0 - 23 )
# busiestDay: Check among all days and return a list of all days which have the maximum number of hours with an 
#             activity.
#
# recurrentActivity: add at a specific hour for all days among a list of days a specific activity (a string).
#              There are three input values:
#              an hour (integer 0-23)
#              a list of days a list [] of numbers between 1 and 365
#              an activity (specified as a string )
# visualiseAgenda: print a simple Cartesian graph where the variable X represent the number of the day and the Y
#                 that of the number of hours with an activity set for that specific day. Remember that in order 
#                 to use the matplotlib library you need to add the command 
#                 "from matplotlib import pylab as plt"
#
# whatToday:  prints the schedule for the day. It takes as an input the number of the day (integer in 1-365)
from matplotlib import pylab as plt

class yearAgenda():
    def __init__(self):
        self.Days = []
        for i in range(365):
            self.Days.append(Day())
        return
    def clearDay(self, day):
        self.Days[day].clearDay()
        return
    def addActivity(self, day, hour, activity):
        self.Days[day].addActivity(hour, activity)
        return
    def removeActivity(self, day, hour):
        self.Days[day].removeActivity(hour)
        return
    def busiestDay(self):
        maxhours = 0
        busiestdays = []
        for i in range(365):
            ihours = self.busyHours(i)
            if ihours > maxhours:
                maxhours = ihours
                busiestdays = [i]
            elif ihours == maxhours:
                busiestdays.append(i)
        print([f"Day {i} is the busiest in the year" for i in busiestdays])
        return busiestdays
    def recurrentActivity(self, hour, daylist, activity):
        for day in daylist:
            self.Days[day].addActivity(hour, activity)
        return
    def busyHours(self, day):
        ihours = 0
        for key in self.Days[day].hours.keys():
            if self.Days[day].hours[key]:
                ihours+=1
        return ihours
    def visualiseAgenda(self):
        y = [self.busyHours(i) for i in range(365)]
        plt.plot(y,'.')
        plt.xlabel('day')
        plt.ylabel('number of busy hours')
        return
    def whatToday(self, day):
        print(f"This is the agenda for day {day}")
        print(self.Days[day].hours)
            
```

```python
thisYear = yearAgenda()

thisYear.addActivity( 300, 9, "Running" )
thisYear.addActivity( 300, 10, "Running" )
thisYear.addActivity( 300, 11, "Running" )
thisYear.addActivity( 300, 12, "Coocking" )

thisYear.addActivity( 321, 9, "Running" )
thisYear.addActivity( 321, 10, "Running" )
thisYear.addActivity( 321, 11, "Running" )
thisYear.addActivity( 321, 12, "Coocking" )

thisYear.whatToday( 321 )

thisYear.recurrentActivity( 15, [ i * 7 for i in range( 52 ) ], "Yoga" )

thisYear.busiestDay()
thisYear.visualiseAgenda()

# If you code it correctly, the following output should appear (or a similar one!), followed by a graph of
# the number of hours vs days which should make sense based on the above!
#This is the agenda for day 321
#Hour 0, activity: None
#Hour 1, activity: None
#Hour 2, activity: None
#Hour 3, activity: None
#Hour 4, activity: None
#Hour 5, activity: None
#Hour 6, activity: None
#Hour 7, activity: None
#Hour 8, activity: None
#Hour 9, activity: Running
#Hour 10, activity: Running
#Hour 11, activity: Running
#Hour 12, activity: Coocking
#Hour 13, activity: None
#Hour 14, activity: None
#Hour 15, activity: None
#Hour 16, activity: None
#Hour 17, activity: None
#Hour 18, activity: None
#Hour 19, activity: None
#Hour 20, activity: None
#Hour 21, activity: None
#Hour 22, activity: None
#Hour 23, activity: None
#Day 300 of the year is among the busiest
#Day 321 of the year is among the busiest
```
