Let us now write a program to solve a generic equation f(x) = 0 using the so-called Newton's bisection algorightm.

This algorithm can be used to solve (almost ) any equation within a certain precision epsilon 
if we know the solution is within an interval [a,b]. 

We will apply it here to solve the equation:

```Python
exp( -alpha *x) - x = 0
```

where `alpha` is a positive constant.  
For x = 0, we have that `exp( -alpha * x ) - x = 1 - 0 > 0`.  
For x = 1 we have `exp( - alpha ) - 1 < 0`.
So we know that the function f(x) = exp( - alpha * x ) - x crosses the zero axis between 0 and 1. Let us now see the algorithm to find the solution!

1. take the mid point between your interval. This is the point x1 = ( a + b ) / 2
2. Calculate f( x1 )
3. if | f( x1 ) | < epsilon, where epsilon is your precision, the program must stop and x1 is your solution, if not:
    - 4.1: if f( x1 ) > 0, then the solution must be within x1 and b ( because f( b ) < 0 ). In this case, set a = x1
    - 4.2: if f( x1 ) < 0 then the solution must be between a and x1. In this case, set b = x1
5. Go back to step 1

Basically, you reiterate between steps 1 and 4 until the condition in Step 3 is verified.

**Help:** the absolute value of a number x is given in Python by writing abs( x ). The exponential function exp( x ) in Python is given by math.exp( x )
