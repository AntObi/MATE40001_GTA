## Exercise 1

Write a program that pics N arbitrary random number from a Gaussian probability distribution, centred in
average = 10 and variance = 2.  
Make a histogram of the N numbers and plot it for different values of $N = 10$,
$10^2$, $10^4$, $10^6$.

````python
import numpy as np
from numpy import random as rnd
from matplotlib import pylab as plt

def pick_gaussians(N, mean, var):
    '''Picks N samples from a Gaussian distribution
    with mean value (mean) and variance (var).'''
    return rnd.normal(mean, np.sqrt(var), size=int(N))

av = 10
var = 2
dist_1 = pick_gaussians(10, av, var)
dist_2 = pick_gaussians(1e2, av, var)
dist_3 = pick_gaussians(1e4, av, var)
dist_4 = pick_gaussians(1e6, av, var)

# plotting
f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4)
f.set_size_inches(18, 5, forward=True)
ax1.hist(dist_1)
ax1.set_title('N=10')
ax2.hist(dist_2, color='green')
ax2.set_title('N=100')
ax3.hist(dist_3, color='orange')
ax3.set_title('N=1e4')
ax4.hist(dist_4, color='darkred')
ax4.set_title('N=1e6')
print(np.var(dist_4))   # -> comparable to input value
````

## Exercise 2

As in the previous exercise, but now the number should be extracted from an exponential distribution of scale parameter = 2.0. You can find the description of the exponential distribution [here](https://numpy.org/doc/stable/reference/random/generated/numpy.random.Generator.exponential.html#numpy.random.Generator.exponential)

````python
def pick_exponentials(N, exp):
    return rnd.exponential(exp, np.int(N))

exp = 2.0
dist_1 = pick_exponentials(10, exp)
dist_2 = pick_exponentials(1e2, exp)
dist_3 = pick_exponentials(1e4, exp)
dist_4 = pick_exponentials(1e6, exp)

f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4)
f.set_size_inches(18, 5, forward=True)
ax1.hist(dist_1)
ax1.set_title('N=10')
ax2.hist(dist_2, color='green')
ax2.set_title('N=100')
ax3.hist(dist_3, color='orange')
ax3.set_title('N=1e4')
ax4.hist(dist_4, color='darkred')
ax4.set_title('N=1e6')
print(np.var(dist_4))   # -> (exp)^2 for sufficent no. of samples
````

## Exercise 3)

1. Write a function myRandom( list1, list2, weights1 = None, weights2 = None ) that takes as input two lists. 
and, optionally, a set of weights (probabilities). The function randomly take one element from the first 
and one element from the second and returns the tuple ( element1, element2 ). Remember that randomly here means that, given a large enough sample, the fraction of times a certain element will be picked is given by `weight[i] / sum_i weight[ i ]`, where `weight[i]` is the probability of extracting element `i`.

2. Using the previous function, write a second function `myRandomArray` that takes as input the two lists and a number NN. The function then generates an array where each element is a tuple previously extracted. 

**Extra part**

3.  Plot the fraction of times each tuple (element1, element2) is extracted as a heat map, using the lists `[ 1, 2, 3, 4, 5 ]` and `[ "a", "b", "c" ]`. A heat map is nothing but a graph where at each coordinate (x,y) a color of a certain intensity (the scale of the heatmap) is associated. For plotting heat-maps, you might want to consult the [link](https://matplotlib.org/3.1.1/gallery/images_contours_and_fields/image_annotated_heatmap.html) as an example.

4. As in 3. above, but now the weights associated to the first list (`[1,2,3,4,5]` ) should be `[0.1, 0.1, 0.1, 0.35, 0.35 ]`, whereas the weights for the second list are all equal.

Question: What do you see comparing the heat maps in 3) and 4)? 


````python

def myRandom( list1, list2, weight1 = None, weight2 = None ):
    '''Picks a touple of two random weighted elemtens from
    list1 and list2.'''
    
    # weighting probabilities
    p1 = (np.ones(len(list1)) / len(list1) if weight1 is None 
          else np.array(weight1) / np.sum(np.array(weight1)))
    p2 = (np.ones(len(list2)) / len(list2) if weight2 is None 
          else np.array(weight2) / np.sum(np.array(weight2)))
    
    # weighted random picks
    x1 = rnd.choice(list1, p=p1)
    x2 = rnd.choice(list2, p=p2)
    
    return ( x1, x2 )

def myRandomArray( list1, list2, NN, weight1 = None, weight2 = None ):
    '''Collects a list of tuples of weighted random elements from
    two lists.'''
    myArray = []    # of type list since input list types variable
    for i in range(NN):
        myArray.append(myRandom(list1, list2, weight1, weight2))
    return myArray

# define lists
list1 = [1, 2, 3, 4, 5]
list2 = ["a", "b", "c"]

# define weights (change to 1's for part 3.)
weight1 = [1, 2, 3, 4, 5]

# obtain list (not an "array" technically) of random tuples
N_trials = 1000
rand_tuples = myRandomArray(list1, list2, N_trials, weight1=weight1)

# some binning to get the list into a plottable format
len1 = len(list1)
len2 = len(list2)
heat_map = np.zeros((len1, len2), dtype=int)

# tuple binning (probably more efficient ways to do this)
for i in range(len(rand_tuples)):
    for k in range(len1):
        for m in range(len2):
            if (rand_tuples[i][0] == list1[k]) and (rand_tuples[i][1] == list2[m]):
                heat_map[k,m] += 1
                break

# plotting (part 3.)
fig, ax = plt.subplots()
im = ax.imshow(heat_map.T)

## from example in the link above
# We want to show all ticks...
ax.set_xticks(np.arange(len(list1)))
ax.set_yticks(np.arange(len(list2)))
# ... and label them with the respective list entries
ax.set_xticklabels(list1)
ax.set_yticklabels(list2)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(list1)):
    for j in range(len(list2)):
        text = ax.text(i, j, "({0}, {1})".format(list1[i], list2[j]),
                       ha="center", va="center", color="w")

ax.set_title("Harvest of local farmers (in tons/year)")
fig.tight_layout()
plt.show()
````

## Exercise 4: Arbitrary probability distribution ( Advanced exercise )

In general, one can extract random number from any (valid) probability distribution following a simple
algorithm, which the exercises here asks you to code.

To do that, you start with a probability distribution $f(x)$ (meaning that $f(x)dx$ is proportional to the probability to extract a number from x to x+dx). $x$ belongs to the interval: $[a,b]$. The algorithm goes like this:

1. Extract a random number "x" between 0 and 1,
2. The random number generated is the number y for which the following condition holds:
   $\int_a^y f(x')dx' / \int_a^b f(x')dx' = x$ (assuming $a\lt y \lt b$).
   
Following the scheme above, write a function `generate( x, y, NN )` that takes an arbitrary probability distribution and returns an array of `NN` values extracted from it. `x` is the array containing the grid points on which $f(x)$ is defined and $y = f(x)$. Note that a valid probability distribution defined over an interval `[a,b]` **must** have the following properties:

a. $\int_a^b f(x)dx < \infty$, or in other words the integral is finite  
b. $f(x)>0 \qquad \forall x$

Experiment with different arbitrary functions and plot the histogram of the generated random numbers (extract a set of N > 10^6 values, using 100 bins.) If done correctly, running the following cell should generate the exponential probability distribution of scale 2.

````python

def generate( x, y, NN ):
    '''Generates NN samples from a given probability distribution
    function.'''
    
    rand_dist = np.zeros(NN)
    
    for i in range(NN):
        x1 = rnd.rand(1)
        
        # compute the integral in [a, b] using the midpoint rule
        int_f = np.trapz(y, x=x)    # careful, y comes before x!
        # do a check for int_f finite if you like here
        
        # iteratively compute the integral up to y divided by int_f
        int_y_div_int = 0.0
        delta_x = x[1] - x[0]    # assuming constant x spacing
        for k in range(1, len(y)):
            y1 = x[k]    # note, the y in the integral is actually an x value!
            if int_y_div_int >= x1:
                # approximated as current y (could do some further linear interpolation 
                # to improve accuracy here)
                break
                
            # repeating np.trapz on [a, y] would work below but is less efficient
            int_y_div_int += 0.5 * (y[k] + y[k-1]) * delta_x / int_f
        rand_dist[i] = y1
    return rand_dist
````

````python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,10**1,10**3) # **5)
y = 1 / 2.0 * np.exp( - x / 2.0 )
s = generate( x, y, 10**4 ) # **6
plt.hist( s, bins = 100)
plt.show()
````

## Exercise 5: Simulating Brownian Dynamics ( Advanced exercise )

Brownian dynamics describes the motion of large particles in a fluid (but, is used also to describe the motion of stock options!). We will see it with a simple example. 

A particle following Brownian dynamics in 1D moves at each step according to the following evolution equation:

$x( t + dt ) = x( t ) + ( 2 D dt )^{1/2} * G(0,1)$

where $G( 0, 1 )$ is a random number generated from a Gaussian of mean zero and variance 1, $dt$ is just a discrete amount of time (called time-step, set it to $10^{-3}$ here), and $D$ the particle diffusion coefficient (in arbitrary units).

For this exercise:

1. Write a function that generates the trajectory of the particle for `NN` timesteps. You can assume x( t = 0 ) = 0

2. Using the function defined in the previous point, write a program that repeat the simulation for `NSample` times, each time restarting from x = 0. You can store the particle trajectory for each of these repeats in a 2D array `traj`, where `traj[ i, j ]` is the position of particle `i` at timestep `j` ( or in other words, at time $t = dt * j$ ). You can store a subset of the values of x every `Nsize` timesteps, rather than for all timesteps simulated.

3. The average displacement after time t is defined as $x_{ave} = \langle x( t ) - x( 0 ) \rangle$ and the mean square displacement as $\sigma^2 = \langle ( x( t ) - x( 0 ) )^2 \rangle$, where $\langle \rangle$ means taking an average over all the `NSample` trajectories generated. By simulation or otherwise, estimate $x_{ave}$ and $\sigma^2$ after evolving the dynamics () for:   
 a) $10^2$ steps and NSample = $10^2$  
 b) $10^4$ steps and NSample = $10^2$  
 c) $10^4$ steps and NSample = $10^4$  
 d) $10^6$ steps and NSample = $10^2$  
 e) $10^6$ steps and NSample = $10^4$  
   
Question: How long does it take to compute all the values for the points b-e? Is there a better way than doing simulations to estimate these quantities?

````python
def generate_trajectory(NN, dt, D):
    '''Generates the particle trajectory for NN timesteps
    using x(t=0) = 0 for given time-step dt and diffusion
    coefficient D.'''
    
    traj = np.zeros(int(NN))
    traj[0] = 0    # initial position
    for i in range(1,int(NN)):
        G = rnd.normal(0.0, 1.0)
        traj[i] = traj[i-1] + np.sqrt(2 * D * dt) * G
    return traj


def generate_set(NSample, NN, dt, D):
    '''Generates NSample trajectories of length NN
    using x(t=0) = 0 for given time-step dt and diffusion
    coefficient D.'''
    
    traj = np.zeros((int(NSample), int(NN)))
    for i in range(int(NSample)):
        traj[i,:] = generate_trajectory(NN, dt, D)
    return traj


def calc_average_displacement(traj):
    '''Calculates the average displacement of N
    particles (rows) simulated for a given time (columns).'''
    
    av_displacement = np.mean(traj, axis=0)    # x0=0; ensemble average (mean) w.r.t. rows (particles)
    return av_displacement


def calc_mean_square_displacement(traj):
    '''Calculates the mean square displacement of N
    particles (rows) simulated for a given time (columns).'''
    
    mean_sq_disp = np.mean(np.square(traj), axis=0)    # x0=0; ensemble average (mean) w.r.t. rows (particles)
    return mean_sq_disp
        
    
# simulation sets
dt = 1e-3
D = 1e-5    # arbitrary value

# a) NN=1e2, NSample = 1e2
NN_a = 1e2
NSample_a = 1e2
time_a = np.linspace(0,int(NN_a)*dt,int(NN_a))
set_a = generate_set(NSample_a, NN_a, dt, D)
mean_a = calc_average_displacement(set_a)
sigma_sq_a = calc_mean_square_displacement(set_a)

f_a, (ax1, ax2) = plt.subplots(1, 2)
f_a.set_size_inches(18, 5, forward=True)
ax1.plot(time_a,mean_a)
ax1.set_xlabel('Time')
ax1.set_ylabel('Mean displacement')
ax2.plot(time_a,sigma_sq_a)
ax2.set_xlabel('Time')
ax2.set_ylabel('Mean square displacement')
plt.show()

# b) NN=1e4, NSample = 1e2
NN_b = 1e4
NSample_b = 1e2
time_b = np.linspace(0,int(NN_b)*dt,int(NN_b))
set_b = generate_set(NSample_b, NN_b, dt, D)
mean_b = calc_average_displacement(set_b)
sigma_sq_b = calc_mean_square_displacement(set_b)

f_b, (ax1, ax2) = plt.subplots(1, 2)
f_b.set_size_inches(18, 5, forward=True)
ax1.plot(time_b,mean_b)
ax1.set_xlabel('Time')
ax1.set_ylabel('Mean displacement')
ax2.plot(time_b,sigma_sq_b)
ax2.set_xlabel('Time')
ax2.set_ylabel('Mean square displacement')
plt.show()

### repeat for c)-e) if you are a patient person...

# The mean displacement average in time approaches zero for sufficient no. of samples.
# The mean square displacement linearly increases in time (on average, for sufficient no. of samples.)
````
